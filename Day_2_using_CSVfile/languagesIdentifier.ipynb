## Main code using dictionary

```python
# Predefined dictionary of programming languages categorized by use case
programming_languages = {
    "General Purpose": ["Python", "Java", "C", "C++", "Ruby", "JavaScript", "Go", "Swift", "Kotlin"],
    "Web Development": ["JavaScript", "PHP", "Ruby", "Python", "TypeScript", "HTML", "CSS"],
    "Systems Programming": ["C", "Rust", "Go", "C++"],
    "Functional Programming": ["Haskell", "Erlang", "Scala", "F#", "Lisp", "Clojure"],
    "Scripting": ["Python", "Ruby", "Perl", "Lua"],
    "Data Science": ["Python", "R", "Julia", "MATLAB"],
    "Mobile Development": ["Swift", "Kotlin", "Java", "Dart"],
    "Game Development": ["C++", "C#", "Java", "Python", "Unity", "UnrealScript"],
    "Statistical Analysis": ["R", "Python", "Julia"],
    "Machine Learning": ["Julia", "Machine Learning"],
    "Data Manipulation and Analysis": ["MySQL", "MATLAB"],
    "Big Data Technologies": ["Big data", "Apache", "Hadoop"],
    "Data Visualization": ["power BI", "tableau", "Matplot"],
    "Deep Learning": ["Deep Learning", "Julia"],
}

def find_common_words(description, programming_languages):
    # Splitting the description into words and converting to lowercase for case-insensitive comparison
    description_words = set(description.lower().split())
    
    # Collecting all words from dictionary values, splitting them into words, and converting to lowercase
    dict_words = set()
    for categories in programming_languages.values():
        for language in categories:
            dict_words.update(language.lower().split())
    
    # Finding common words between the dictionary values and the description
    common_words = dict_words.intersection(description_words)
    
    return common_words

# Prompting user to input a description
user_description = input("Enter your description: ")

# Finding and displaying common words
common_words = find_common_words(user_description, programming_languages)
print("Common words between the description and programming languages:", common_words)
```

### Displaying the tecnical words in resume

```python
import csv
from docx import Document
import re

def read_docx_remove_numbers_special_chars(doc_path):
    doc = Document(doc_path)
    words_cleaned = []

    for para in doc.paragraphs:
        words = re.findall(r'\b[a-zA-Z]+\b', para.text)
        words_cleaned.extend(word.lower() for word in words)

    # Remove duplicates while preserving order
    words_list = list(dict.fromkeys(words_cleaned))

    return words_list

# Specify the path to your Word document
doc_path = 'resume.docx'

# Call the function and get the list of words without numbers and special characters, all in lowercase
words_list = read_docx_remove_numbers_special_chars(doc_path)

def find_technologies_in_description(description, technologies):
    common_technologies = set()
    
    # Checking for technologies in the description
    for technology in technologies:
        if technology.lower() in description:
            common_technologies.add(technology)
    
    return common_technologies

# Prompting user for description
# user_description = input("Enter your description: ")

# Finding and displaying common technologies
common_technologies_myresume = find_technologies_in_description(words_list, technologies_list)
print("Technologies mentioned:", common_technologies_myresume)
```

### Display the matched key words and %

```python
import csv

common_technologies_myresume = ['analysis', 'powerpoint', 'ai', 'reporting', 'seaborn', 'python', 'system', 'pandas', 'javascript', 'adobe', 'php', 'gmail', 'numpy', 'engineering', 'technical', 'engagement', 'excel', 'content', 'java', 'queries', 'sql', 'js', 'oracle', 'analytics', 'tableau', 'r', 'reports', 'database', 'schedule', 'matplotlib', 'word', 'mysql', 'mining', 'programming', 'c', 'etl', 'metrics']
csv_file_path = r'C:\Users\Priyaa\Documents\GitHub\Language_identification_python\skills.csv'

# Initialize an empty list to store your data
data_list = []

# Open the CSV file and read the data
with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        data_list.append(row)

technologies_list = [item for sublist in data_list for item in sublist]

def find_technologies_in_description(description, technologies):
    description_lower = description.lower()
    common_technologies = set()
    
    # Checking for technologies in the description
    for technology in technologies:
        if technology.lower() in description_lower:
            common_technologies.add(technology)
    
    return common_technologies

# Prompting user for description
user_description = input("Enter your description: ")

# Finding and displaying common technologies
common_technologies = find_technologies_in_description(user_description, technologies_list)

def calculate_match_percentage(words, technologies):
    # Convert all words to lowercase for case-insensitive matching
    words_lower = [word.lower() for word in words]
    technologies_lower = [tech.lower() for tech in technologies]

    # Find common technologies in the description
    common_technologies = set(tech for tech in technologies_lower if tech in words_lower)
    
    # Calculate the match percentage
    match_percentage = (len(common_technologies) / len(set(technologies))) * 100 if technologies else 0
    
    return common_technologies, match_percentage

# Calculate match percentage based on the resume words and technologies
common_technologies, match_percentage = calculate_match_percentage(common_technologies_myresume, common_technologies)

print("Technologies mentioned:", common_technologies)
print("Match percentage:", match_percentage)
```
